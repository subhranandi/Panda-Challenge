## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
a = school_data_complete.groupby(['school_name'])['school_name','type'].max()

#b = school_data_complete.groupby(['school_name'])['school_name','type','grade'].max()

ss = pd.DataFrame(school_data_complete.groupby(['school_name','type'])['school_name','type','budget','size'].mean())

# per student budget
x = pd.DataFrame(ss['budget']/ss['size'])

# counting Average Math Score & Average Reading Score

m = pd.DataFrame(school_data_complete.groupby(['school_name','type'])['school_name','type','math_score','reading_score'].mean())

# counting % Passing Math
pass_math = school_data_complete[school_data_complete['math_score']>=70]
all_math = school_data_complete[school_data_complete['math_score']>=0]

pm = pd.DataFrame(pass_math.groupby(['school_name','type'])['math_score'].count())
am = pd.DataFrame(all_math.groupby(['school_name','type'])['math_score'].count())
fullpm = pd.concat([pm,am],axis=1)
fullpm.columns = ['passcount','allcount']
xx = pd.DataFrame(fullpm['passcount']/fullpm['allcount']*100)

# counting % passing reading
pass_read = school_data_complete[school_data_complete['reading_score']>=70]
all_read = school_data_complete[school_data_complete['reading_score']>=0]

pr = pd.DataFrame(pass_read.groupby(['school_name','type'])['reading_score'].count())
ar = pd.DataFrame(all_read.groupby(['school_name','type'])['reading_score'].count())
fullpr = pd.concat([pr,ar],axis=1)
fullpr.columns = ['passcount','allcount']
xy = pd.DataFrame(fullpr['passcount']/fullpr['allcount']*100)

#counting Overall Passing Rate (Average of the above two)
opr = (xx+xy)/2



y = pd.concat([a.reset_index(drop=True),ss.reset_index(drop=True),x.reset_index(drop=True),m.reset_index(drop=True),xx.reset_index(drop=True),xy.reset_index(drop=True),opr.reset_index(drop=True)],axis=1)
y.columns = ['School Name','Type','Budget','Total Students','Per Student Budget','Avg Math Score','Avg Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate']
y

```
